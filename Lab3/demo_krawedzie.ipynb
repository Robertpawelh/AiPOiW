Przykłady detekcji krawędzi (Sobel, Canny) w OpenCV

```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image
```

Przykład 1: Lena

```python
lena = cv2.imread('Lena.jpg', cv2.IMREAD_GRAYSCALE)
Image.fromarray(lena)
```

Detekcja krawędzi metodą Sobela

```python
#Detekcja krawędzi metodą Sobela - w kierunku X:
sobelx = cv2.Sobel(src=lena, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=3)

#Detekcja krawędzi metodą Sobela - w kierunku Y:
sobely = cv2.Sobel(src=lena, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=3)

#Detekcja krawędzi metodą Sobela - w kierunkach X i Y:
sobelxy = cv2.Sobel(src=lena, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=3)

f, (a1, a2, a3) = plt.subplots(1, 3)
f.set_figwidth(15)
f.set_figheight(5)
a1.imshow(sobelx, cmap='gray');
a1.set_title('krawędzie o orientacji pionowej');
a2.imshow(sobely, cmap='gray');
a2.set_title('krawędzie o orientacji poziomej');
a3.imshow(sobelxy, cmap='gray');
a3.set_title('krawędzie o orientacjach pionowej i pionowej');
```

Detekcja krawędzi metodą Canny'ego

```python
#Detekcja krawędzi metodą Canny'ego - wynik zależy od doboru wartości progów:
f, ((a1, a2, a3), (a4, a5, a6)) = plt.subplots(2, 3)
f.set_figwidth(15)
f.set_figheight(10)

a1.imshow(lena, cmap='gray');
a1.set_title('obraz oryginalny')

for ax, t1, t2 in zip([a2, a3, a4, a5, a6], [50, 50, 250, 20, 20], [200, 300, 300, 100, 300]):
    ax.imshow(cv2.Canny(image=lena, threshold1=t1, threshold2 = t2), cmap='gray');
    ax.set_title('Canny, próg dolny = {}, próg górny = {}'.format(t1, t2))
```

Przykład 2: monety

```python
coins = cv2.imread('Monety_kolor.bmp', cv2.IMREAD_GRAYSCALE)
Image.fromarray(coins)
```

Detekcja krawędzi metodą Sobela

```python
sobelx = cv2.Sobel(src=coins, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=3)
sobely = cv2.Sobel(src=coins, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=3)
sobelxy = cv2.Sobel(src=coins, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=3)

f, (a1, a2, a3) = plt.subplots(3, 1)
f.set_figwidth(25)
f.set_figheight(20)
a1.imshow(sobelx, cmap='gray');
a1.set_title('krawędzie o orientacji pionowej');
a2.imshow(sobely, cmap='gray');
a2.set_title('krawędzie o orientacji poziomej');
a3.imshow(sobelxy, cmap='gray');
a3.set_title('krawędzie o orientacjach pionowej i pionowej');
```

Detekcja krawędzi metodą Canny'ego

```python
f, (a1, a2, a3) = plt.subplots(3, 1)
f.set_figwidth(25)
f.set_figheight(20)

a1.imshow(lena, cmap='gray');
a1.set_title('obraz oryginalny')

for ax, t1, t2 in zip([a1, a2, a3], [20, 50, 250], [50, 300, 500]):
    ax.imshow(cv2.Canny(image=coins, threshold1=t1, threshold2 = t2), cmap='gray');
    ax.set_title('Canny, próg dolny = {}, próg górny = {}'.format(t1, t2))
```
