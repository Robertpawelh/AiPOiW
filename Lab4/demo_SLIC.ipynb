```python
import cv2
from PIL import Image
from skimage.segmentation import slic, mark_boundaries
from skimage.util import img_as_float
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('Lena.jpg',cv2.IMREAD_COLOR)
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
Image.fromarray(img_RGB)
```

```python
image = img_as_float(img_RGB)

# Pętla po liczbie klastrów:
for numSegments in (100, 300, 600):
  segments = slic(image, n_segments=numSegments, sigma=5, start_label=1)

  fig = plt.figure(figsize=(10, 10))
  ax = fig.add_subplot(1,1,1)
  img_tmp = mark_boundaries(image, segments)
  ax.imshow(img_tmp) 
  plt.axis("off")
```
